## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Before doing any data wrangling lets create copy of the dataset
df2 = df.copy()
```

```python
# Checking the number of rows with zero price values in our dataset.
df2[df2.price==0].shape
```

Here we can clearly see that these eleven entries need to be drop in order to get meaningful analysis so we will be dropping these entries where price equals to zero ("0").

```python
# Removing rows having zero price values from our dataset.
df2 = df2[df2['price'] !=0]
```

The columns with missing values are name, host_name, last_review, and reviews_per_month - with the last two missing over 10,000 rows of data each.  
The number of reviews per month can be interesting to analyze, so the best way to deal with this column is to replace all NaN values with zeros, similarly we can host_names and names are not that important in our analysis, so atleast we are good to fill those with some substitutes in both the columns! columns, they will not be used in this analysis and should be dropped.

```python
# Since both 'name' and 'host_name' has very less missing values as compared to the 
# entire dataset we can fill these missing values in both with unknown & no_name respectively!

df2['name'].fillna('unknown',inplace=True)
df2['host_name'].fillna('anonymous',inplace=True)
```

```python
# Checks if any null values still exist!
df2[['host_name','name']].isnull().values.any() 
```

```python
# Filling null values in reviews_per_month column with 0
df2['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

```python
# Drop the last_review column
df2.drop(['last_review'], axis=1, inplace=True)
```

```python
# Checking change in shape
df2.shape
```

Previous shape was (48895, 16) and the shape of the new data frame created after cleaning is (48884, 15)
Hence correctly removed 11 rows where the price value was 0

## Handling the outliers

Outliers in the price column

```python
sns.boxplot(df2['price'])
```

Removing Outliers using IQR approach! (has performed best in removing outliers)

```python
# Calculate the IQR and boundaries
Q1 = df2['price'].quantile(0.25)
Q3 = df2['price'].quantile(0.75)
IQR = Q3 - Q1
lower_boundary = Q1 - 1.5 * IQR
upper_boundary = Q3 + 1.5 * IQR

# Filter the dataset to remove outliers
df2 = df2[(df2['price'] >= lower_boundary) & (df2['price'] <= upper_boundary)]
```

```python
sns.boxplot(df2['price']) #the boxplot for price now looks good without outliers! 
```

```python
sns.distplot(df2['price'])
```

Outliers in the minimum nights column

```python
# Calculate the IQR and boundaries
Q1 = df2['minimum_nights'].quantile(0.25)
Q3 = df2['minimum_nights'].quantile(0.75)
IQR = Q3 - Q1
lower_boundary = Q1 - 1.5 * IQR
upper_boundary = Q3 + 1.5 * IQR

# Filter the dataset to remove outliers
df2 = df2[(df2['minimum_nights'] >= lower_boundary) & (df2['minimum_nights'] <= upper_boundary)]
```

```python
sns.boxplot(df2['minimum_nights']) #the boxplot for min nights now looks good without outliers! 
```

```python
sns.distplot(df2['minimum_nights'])
```

### Univariate

#### Chart - 1

```python
# Chart - 1 visualization code

for col in df.describe().columns:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))
    
    # Histogram plot
    sns.histplot(df2[col], ax=axes[0], kde=True, color='skyblue', edgecolor='white')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frequency')
    axes[0].set_title("Distribution of " + col, fontsize=15)
    axes[0].grid(True, linestyle='--')
    
    # Boxplot
    sns.boxplot(df2[col], ax=axes[1], orient='h', showmeans=True, color='lightgreen')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('')
    axes[1].set_title("Boxplot of " + col, fontsize=15)
    axes[1].grid(True, linestyle='--')
    
    # Adjust spacing between subplots
    plt.tight_layout()
    
    # Show the plots
    plt.show()
```

##### 1. Why did you pick the specific chart?

A histplot is a type of chart that displays the distribution of a dataset. It is a graphical representation of the data that shows how often each value or group of values occurs. Histplots are useful for understanding the distribution of a dataset and identifying patterns or trends in the data. It is also useful when dealing with large data sets (greater than 100 observations). It can help detect any unusual observations (outliers) or any gaps in the data.

Thus, I used the histogram plot to analysis the variable distributions over the whole dataset whether it's symmetric or not.

A boxplot is used to summarize the key statistical characteristics of a dataset, including the median, quartiles, and range, in a single plot. Boxplots are useful for identifying the presence of outliers in a dataset, comparing the distribution of multiple datasets, and understanding the dispersion of the data. They are often used in statistical analysis and data visualization.

Thus, for each numerical varibale in the given dataset, I used box plot to analyse the outliers and interquartile range including mean, median, maximum and minimum value.

#### What is the average price for each neighbourhood group?

```python
# Group the dataset by "Neighbourhood group" and calculate the average price
neighborhood_avg_price = df2.groupby('neighbourhood_group')['price'].mean()

# Set the style of the plot
sns.set(style="whitegrid")

# Create a bar plot with custom color palette
plt.figure(figsize=(10, 6))
colors = sns.color_palette('pastel')
ax = sns.barplot(x=neighborhood_avg_price.index, y=neighborhood_avg_price, palette=colors)

# Set the title and axis labels
plt.title('Neighbourhood Group vs. Average Price', fontsize=16)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=30)

# Add value labels on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()),ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)
  
# Remove the top and right spines
sns.despine()
    
#display the plot    
plt.tight_layout()
plt.show()
```

#### Identifying the top 30 busiest Airbnb hosts based on average review rates and exploring the distribution of review rates across different room types for these hosts.

```python
# Chart - 12 visualization code

# Create a series of top 30 hosts by review rate
review_rate = df2.groupby('host_id')['reviews_per_month'].mean().head(30).reset_index()
review_rate.sort_values('reviews_per_month', ascending=False, inplace=True)
review_rate = review_rate.iloc[:, 0]

# Create a new dataframe containing information of only these top 30 hosts
Top_host_by_review_rate = df2.loc[df2['host_id'].isin(review_rate)]
Top_host_by_review_rate = Top_host_by_review_rate.sort_values('reviews_per_month')

# Create a catplot to visualize the review rates of these top 30 hosts by room type
fig = sns.catplot(data=Top_host_by_review_rate, x='host_id', y='reviews_per_month',col='room_type', kind='bar', height=2, aspect=8)
fig.set_xticklabels(rotation=90)
fig.fig.set_size_inches(20, 10)
```
